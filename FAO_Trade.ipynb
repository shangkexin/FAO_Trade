```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
# download data from FAO
df = pd.read_csv('Trade_CropsLivestock_E_All_Data_(Normalized)/Trade_CropsLivestock_E_All_Data_(Normalized).csv',encoding='windows-1252')
trade_agg = pd.read_csv('Trade_CropsLivestock_E_All_Data_(Normalized)/Trade_CropsLivestock_E_ItemCodes.csv',encoding='windows-1252')
# the downloaded data has been modified by adding Item Group Columns
trade_agg = trade_agg[~trade_agg['Item Group'].isnull()]
df = df[df['Unit'] != 'No']
df = df[df['Item Code'].isin(trade_agg['Item Code'])]
df
```

```python
trade = pd.read_excel('DataDict.xlsx')
trade
```

```python
# download from WDI
gdp = pd.read_excel('API_NY.GDP.MKTP.CD_DS2_en_excel_v2_42476.xls', 
                    sheet_name='Data',
                    skiprows=3)
gdp = gdp.drop(columns=['Country Code',
                        'Indicator Name',
                        'Indicator Code'])
country_wdi = pd.read_excel('Country.xlsx',
                            sheet_name= 'WDI')

gdp = gdp.set_index('Country Name')
gdp = pd.DataFrame(gdp.stack(dropna = False), 
                  columns = ['gdp']).reset_index().rename(columns={'level_1':'Year'})
gdp = country_wdi.merge(gdp, 
                        # left_on='Country_WDI',
                        on='Country Name',
                        how = 'left')
gdp = gdp.drop(columns=['Country Name'])
gdp.gdp = gdp.gdp.apply(pd.to_numeric, errors='coerce')
gdp.Year = gdp.Year.apply(pd.to_numeric, errors='coerce')
gdp
```

```python
country = pd.read_excel('Country.xlsx')
```

```python
for group in trade_agg['Item Group'].unique():
    dd = trade[trade['Item Group'] == group]
    agg = trade_agg[trade_agg['Item Group'] == group]
    dt = df[df['Item Code'].isin(agg['Item Code'])]
    for j in dd.index:
        # if 'Value' not in dd.loc[j,'Variable']:
        v = dd.loc[j,'Variable']
        m = int(dd.loc[j,'Element Code'])
        if dd.isnull().loc[j,'Decimal Places']:
            d = dd.loc[j,'Decimal Places']
        else:
            d = int(dd.loc[j,'Decimal Places'])
        dt1 = dt[dt['Element Code'] == m]
        dt1 = dt1[['Area','Year','Value']]
        dt1.Year = dt1.Year.apply(pd.to_numeric, errors='coerce')
        dt1 = dt1.groupby(['Area','Year']).sum(min_count=1).reset_index()
        if 'Value' not in dd.loc[j,'Variable']:
            print(v)
            if dd.isnull().loc[j,'Decimal Places']:
                dt1 = dt1
            else:
                dt1.Value = round(dt1.Value, d)
        else:
            print(v)
            dt1.Value = dt1.Value*1000
            dt1 = country.merge(dt1,
                    on = 'Area',
                    how = 'left')
            dt1 = dt1.merge(gdp,
                            on = ['Country', 'FIPS_CODE','Year'],
                            how = 'left'
                            )
            if dd.isnull().loc[j,'Decimal Places']:
                dt1 = dt1
            else:
                dt1['val'] = round(dt1['Value']/dt1['gdp']*100,d)
            dt1 = dt1[['Area','Year','val']].rename(columns={'val':'Value'})

        dt1 = country.merge(dt1,
                        on = 'Area',
                        how = 'left')
        dt1.drop(columns='Area', inplace = True)

        first_last = dt1.sort_values(['Country','Year']).groupby(dt1['Country'])['Value'] \
            .agg(['first', 'last']).reset_index()
        first_last = first_last.rename(columns={"first": "Earliest", "last": "MostRecent"})
        dt1.Year = dt1.Year.astype("Int64")

        dt1 = dt1.groupby(['Country','Year']).sum(min_count=1).reset_index()

        dt1 = dt1.pivot_table(index = 'Country',
                    columns='Year',
                    values = 'Value').reset_index()
        dt1 = dt1.merge(first_last,
                        on = 'Country',
                        how = 'left'
                        )
        dt1 = country.merge(dt1,
                        on = 'Country',
                        how = 'left')
        dt1.drop(columns='Area', inplace = True)
        dt1.dropna(axis = 0, how = 'all', inplace=True)
        dt1 = dt1.drop_duplicates(ignore_index=True)
        dt1 = dt1.sort_values(by='Country')
        dt1 = dt1.set_index(['Country','FIPS_CODE'])
        dt1 = dt1.apply(lambda x: pd.Series.round(x, d))
        # dt1.to_csv(f'IFs Import/{v}.csv',
        #             #  index=False
        #              )
dt1             

```
